```python
df['Memory'] = df['Memory'].astype(str).replace(r'\.0', '', regex=True)
df['Memory'] = df['Memory'].str.replace('GB', '')
df['Memory'] = df['Memory'].str.replace('TB', '000')
new = df['Memory'].str.split('+',n=1, expand=True)

df['first'] = new[0]
df['first'] = df['first'].str.strip()

df['second'] = new[1]

df['Layer1HDD'] = df['first'].apply(lambda x: 1 if 'HDD' in x else 0)
df['Layer1SSD'] = df['first'].apply(lambda x: 1 if 'SSD' in x else 0)
df['Layer1Hybrid'] = df['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df['Layer1Flash_Storage'] = df['first'].apply(lambda x: 1 if 'Flash Strorage' in x else 0)

df['first'] = df['first'].str.replace(r'\D', '',regex=True)

df['second'] = df['second'].fillna('0')

df['Layer2HDD'] = df['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df['Layer2SSD'] = df['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df['Layer2Hybrid'] = df['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)
df['Layer2Flash_Storage'] = df['second'].apply(lambda x: 1 if 'Flash Strorage' in x else 0)

df['second'] = df['second'].str.replace(r'\D', '',regex=True)

df['first'] = df['first'].astype(int)
df['second'] = df['second'].astype(int)

df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])

df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

## Voting Regressor

```python
from sklearn.ensemble import VotingRegressor

step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 10, 11])
], remainder='passthrough')

# Define the regressors
rf = RandomForestRegressor(n_estimators=350, random_state=3,bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=15)
gbdt = GradientBoostingRegressor(n_estimators=100, max_features=0.5)
et = ExtraTreesRegressor(n_estimators=100, random_state=3,bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=10)

# Step 2: Voting Regressor
step2 = VotingRegressor([('rf', rf), ('gbdt', gbdt), ('et', et)], weights=[5, 1, 1])

# Build the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline
pipe.fit(X_train, Y_train)

# Predict
y_pred = pipe.predict(X_test)

# Evaluate
print('R2 score:', r2_score(Y_test, y_pred))
print('MAE:', mean_absolute_error(Y_test, y_pred))
```

## Stacking

```python
from sklearn.ensemble import StackingRegressor, RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_absolute_error

# Define the estimators, including ExtraTreesRegressor
estimators = [
    ('rf', RandomForestRegressor(n_estimators=350, random_state=3,bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=15)),
    ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
    ('et', ExtraTreesRegressor(n_estimators=100, random_state=3,bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=10))
]

# Define the stacking regressor with Ridge as the final estimator
step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit the pipeline
pipe.fit(X_train, Y_train)

# Predict using the trained model
y_pred = pipe.predict(X_test)

# Print performance metrics
print('R2 score', r2_score(Y_test, y_pred))
print('MAE', mean_absolute_error(Y_test, y_pred))
```

```python
from sklearn.ensemble import StackingRegressor, RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor, VotingRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.pipeline import Pipeline

# Define the base estimators for VotingRegressor
voting_regressor = VotingRegressor([
    ('rf', RandomForestRegressor(n_estimators=350, random_state=3, bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=15)),
    ('gbdt', GradientBoostingRegressor(n_estimators=100, max_features=0.5)),
    ('et', ExtraTreesRegressor(n_estimators=100, random_state=3, bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=10))
])

# Define other estimators for stacking
estimators = [
    ('voting', voting_regressor),  # Use VotingRegressor as one of the base models
    ('rf', RandomForestRegressor(n_estimators=350, random_state=3, bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=15)),
    ('et', ExtraTreesRegressor(n_estimators=100, random_state=3, bootstrap=True, max_samples=0.5, max_features=0.75, max_depth=10))
]

# Define the stacking regressor with Ridge as the final estimator
step2 = StackingRegressor(estimators=estimators, final_estimator=Ridge(alpha=100))

# Create the pipeline
pipe = Pipeline([
    ('step1', step1),  # Assuming step1 is your preprocessing pipeline
    ('step2', step2)
])

# Fit the pipeline
pipe.fit(X_train, Y_train)

# Predict using the trained model
y_pred = pipe.predict(X_test)

# Print performance metrics
print('R2 score', r2_score(Y_test, y_pred))
print('MAE', mean_absolute_error(Y_test, y_pred))
```
